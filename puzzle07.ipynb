## Finding products which have color variations

```python
colors = ["red", "green", "blue", "magenta", "purple", "amber", "azure", 
           "white", "yellow", "mauve", "puce", "orange", "dreidel"]
```

```python

grouped_products = products.copy()
grouped_products["desc"] = grouped_products["desc"].str.replace("|".join(colors), "", case=False, regex=True)
grouped_products["desc"] = grouped_products["desc"].str.replace("\(|\)", "", regex=True)
grouped_products = grouped_products[["desc","sku"]].groupby("desc").aggregate("|".join)
grouped_products = grouped_products[grouped_products["sku"].str.contains("\|", regex=True)]
grouped_products
```

## Filtering for items with color variations

```python
colored_skus = set("|".join(grouped_products["sku"]).split("|"))

orders_items = orders_items[orders_items["sku"].isin(colored_skus)]
orders_items = orders_items.merge(products, on="sku").drop(columns=["qty", "wholesale_cost","unit_price"])

orders_items["category"] = orders_items["desc"]
orders_items["category"] = orders_items["category"].str.replace("|".join(colors), "", case=False, regex=True)
orders_items["category"] = orders_items["category"].str.replace("\(|\)", "", regex=True)
orders_items
```

```python
orders_items = orders_items.merge(orders, on="orderid")
orders_items.drop(columns=["items", "total"], inplace=True)
orders_items = orders_items.merge(customers[["customerid","name","phone"]], on="customerid")

orders_items
```

### Emily's orders

```python
emily_orders = orders[orders['customerid']==emily_id]
emily_orders
```

```python
emily_orders_items = orders_items[orders_items['customerid']==emily_id]
emily_orders_items
```

## Finding a pair for Emily's orders
- same category
- pair of orders close in time

```python
from IPython.display import display

t_cutoff = 600 # in seconds

for idx, emily_order in emily_orders_items.iterrows():
    orders_items["closeby"] = False
    shipped_date = emily_order["shipped"]
    seconds_filter = abs((orders_items["shipped"] - shipped_date).dt.total_seconds()) <= t_cutoff
    orders_items.loc[seconds_filter, "closeby"] = True
    cat_filter = (orders_items["category"] == emily_order["category"])
    matches = orders_items[(orders_items["closeby"]) & (cat_filter) ]
    
    if len(matches)>1:
        print(60*"-")
        print(f"Matches on date {shipped_date}")
        display(matches)
```
