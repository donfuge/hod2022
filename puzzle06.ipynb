```python
customers, orders, orders_items, products = load_data()
```

Looking for sales where the unit price is smaller than the wholesale price


```python
orders_items = orders_items.merge(products, on="sku")
orders_items = orders_items[orders_items["unit_price"]<orders_items["wholesale_cost"]]

orders_items
```

```python
orders = orders[orders['ordered'] >= '2017-04-05']
orders_items = orders_items.merge(orders, on="orderid")
orders_items
```

```python
counts = orders_items["customerid"].value_counts()
counts
```

Customer with the most purchases under wholesale price

```python
customers[customers["customerid"]==counts.index[0]]
```
