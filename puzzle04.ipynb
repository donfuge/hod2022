```python
import pandas as pd
from utils import load_data
```

```python
customers, orders, orders_items, products = load_data()
```

## Looking for pastry products
Idea: check for items with flavor variations.


```python
flavors = ["raspberry", "poppy", "sesame", "caraway", "kasha", "knish"]
pastries = products[products["desc"].str.contains("|".join(flavors), case=False)]
pastry_skus = set(pastries["sku"])
pastries
```

## Filtering orders
- pastry products
- ordered between 4am and 5am
- ordered after 2017-04-05

```python
orders
```

```python
orders = orders[orders['ordered'] >= '2017-04-05']
orders = orders[orders['ordered'].dt.hour < 5]
orders = orders[orders['ordered'].dt.hour >= 4]

orders  
```

```python
orders = orders.merge(orders_items[["orderid","sku"]], on="orderid")
orders = orders[orders["sku"].isin(pastry_skus)]
orders
```

```python
orders = orders.merge(products[["sku","desc"]], on="sku")
orders = orders.merge(customers, on="customerid").drop(columns=["shipped","total","address","items"])
orders
```

## Ranking customers

```python
counts = pd.DataFrame(orders["customerid"].value_counts().rename("counts"))
counts.index.name="customerid"
counts
```

```python
counts = counts.merge(customers, on="customerid")
counts
```
